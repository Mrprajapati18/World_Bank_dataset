```python
Health.rename(columns={'Country_Region': 'country'}, inplace=True)
Health.rename(columns={'Province_State' : 'province'}, inplace = True)
Health.rename(columns={'World_Bank_Name': 'world_bank_name'}, inplace = True)
Health.rename(columns={'Health_exp_pct_GDP_2016' : 'expenditure_percentage_gdp'}, inplace = True)
Health.rename(columns={'Health_exp_public_pct_2016' : 'expenditure_percentage_public'}, inplace = True)
Health.rename(columns={'Health_exp_out_of_pocket_pct_2016' : 'expenditure_percentage_pocket'}, inplace = True)
Health.rename(columns={'Health_exp_per_capita_USD_2016' : 'per_capita_dollar'}, inplace = True)
Health.rename(columns={'per_capita_exp_PPP_2016' : 'per_capita_dollar_ppp'}, inplace = True)
Health.rename(columns={'External_health_exp_pct_2016' : 'external_money_percentage'}, inplace = True)
Health.rename(columns={'Physicians_per_1000_2009-18' : 'physicians_per_1000'}, inplace = True)
Health.rename(columns={'Nurse_midwife_per_1000_2009-18' : 'nurse_midwife_per_1000'}, inplace = True)
Health.rename(columns={'Specialist_surgical_per_1000_2008-18' : 'specialist_per_100000'}, inplace = True)
Health.rename(columns={'Completeness_of_birth_reg_2009-18' : 'birth_percentage'}, inplace = True)
Health.rename(columns={'Completeness_of_death_reg_2008-16' : 'death_percentage'}, inplace = True)
```

# Box ploting

# Heatmap

```python
# Exclude non-numeric columns from the correlation matrix
numeric_health = Health.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr_matrix = numeric_health.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Create a heatmap with seaborn
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", vmin=-1, vmax=1)  

# Show the plot
plt.title('World Bank dataset')
plt.show()
```
